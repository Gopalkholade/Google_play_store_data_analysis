```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_excel('googleplaystore.xlsx')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
impute=SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
impute.fit(df.iloc[:,2:3].values)
```

```python
df.iloc[:,2:3]=impute.transform(df.iloc[:,2:3].values)
```

```python
df.isnull().sum()
```

```python
 df =df.dropna()
```

```python
df.isnull().sum()
```

```python
df.head(2)
```

## How many free apps are there in art and design

```python
dn=df[(df['Category']=='ART_AND_DESIGN')]
len(dn[dn['Type']=='Free'])
```

## How many apps are there in Family cate. with rating greater 4.5  and free as well 

```python
dn1=df[df['Category']=='FAMILY']
dn2=dn1[dn1['Rating']>4.5]
len(dn2[dn2['Type']=='Free'])
```

## list all the free apps where rating is more than 4.5 and categorie is family

```python
df2=df[df['Type']=='Free']
df3=df2[df2['Rating']>4.5]
d=df3[df3['Category']=="FAMILY"]['App'].values
```

```python
c=0
for i in d:
    print(i)
    c+=1
print(c)
```

### Name the apps that are in ART_and_Design with rating more than 4.5 in ascending order WRT ratings 

```python
dn=df[(df['Category']=='ART_AND_DESIGN')]
sd=dn[dn['Type']=='Free']
sd.sort_values(by='Rating',ascending=True)
```

### Name the apps that are in ART_and_Design and free  in decending order WRT review 

```python
dn=df[(df['Category']=='ART_AND_DESIGN')]
sd=dn[dn['Type']=='Free']
sd.sort_values(by='Reviews',ascending=False)
```

### which category is having maximum average rating

```python
df.groupby('Category').mean()['Rating'].sort_values(ascending=False)
```

### How many free apps are there in each category in descending Order?

```python
df_pr=df[df['Type']=='Paid']
df_pr.groupby('Category').count()['Type'].sort_values(ascending=False)
```
